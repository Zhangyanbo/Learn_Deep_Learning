```python
import torch as torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import pandas as pd
import math
import torch.optim as optim
```

## 数据导入与训练

```python
raw_data = pd.read_csv("hour.csv")[1:600]
#raw_data[:10]

def normalize(data):
    _mean = np.mean(data)
    _std = np.std(data)
    
    return (data - _mean) / _std

# One-hot
def one_hot(num, total, bias = 0):
    _list = [0] * total
    _list[num - bias] = 1
    return _list

def make_input(database, index):
    _datalist = []
    for key in ['season', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']:
        _datalist += database[key][index]
        
    _datalist += [database['temp'][index], database['atemp'][index], database['hum'][index]]
    return torch.Tensor(_datalist).cuda()

# 数据预处理
transfored_data = raw_data

for key in ['cnt', 'registered', 'hum', 'atemp', 'temp']:
    transfored_data[key] = normalize(transfored_data[key])

#transfored_data[1:10]

for key, total, _bias in [['workingday', 2, 0], ['weathersit', 4, 1], ['weekday', 7, 0], \
                          ['holiday', 2, 0], ['hr', 24, 0], ['mnth', 12, 1], ['season', 4, 1]]:
    transfored_data[key] = [one_hot(_num, total, bias = _bias) for _num in transfored_data[key]]

#训练

import random

net = nn.Sequential(nn.Linear(34, 34), nn.ReLU(), nn.Linear(34, 10), nn.ReLU(), \
                    nn.Linear(10, 10), nn.ReLU(), nn.Linear(10, 1))

net.to(torch.device('cuda:0'))

criterion = nn.MSELoss()

batchSize = 2
#learning_rate = 0.01
meanLoss = 0
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)

running_loss = 0.0
t = 0

for k in range(1000):
    meanLoss = 0
    
    #running_loss = 0.0
    optimizer.zero_grad()
    indexs = list(range(1,500))
    random.shuffle(indexs)
    for i in indexs:
        t += 1
        # Training
        optimizer.zero_grad()
        input = make_input(transfored_data, i)
        output = net(input)
        loss = criterion(output, torch.Tensor([transfored_data['cnt'][i]]).cuda())
        
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        #if t < 20:
        #    print('start loss: ', loss.item())
        if t % 6400 == 0:
            print('loss: ', running_loss / 6400.)
            running_loss = 0.0
            
        
    #print('Trun', k)
    
print('finished')
```

```python
maxt = 600

result_compare=pd.DataFrame({'t':[i for i in range(1, maxt)] + [i for i in range(1, maxt)], \
              'data':[net(make_input(transfored_data, i)).item() for i in range(1, maxt)] \
                       + [transfored_data['cnt'][i] for i in range(1, maxt)],\
                'type':['predict' for i in range(1, maxt)] + ['real' for i in range(1, maxt)]
             })

result_compare[1:10]
```

## 简单的$\sin$函数实验

```python
import numpy as np
import torch as torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

traindata = torch.Tensor([[[x.item()], [np.sin(4 * 3.1415926 * x.item())]] for x in torch.rand(3000)])
print(traindata[:10])
```

```python
net = nn.Sequential(nn.Linear(1, 10), nn.Sigmoid(), nn.Linear(10, 10), nn.Sigmoid(), nn.Linear(10, 1))
print(net)
```

```python
net = nn.Sequential(nn.Linear(1, 5), nn.Sigmoid(), nn.Linear(5, 10), nn.Sigmoid(), \
                    nn.Linear(10, 10), nn.Sigmoid(), nn.Linear(10, 1))

criterion = nn.MSELoss()
optimizer = optim.SGD(net.parameters(), lr=0.01, momentum=0.9)

inputs = traindata.transpose(0,1)[0]
targets = traindata.transpose(0,1)[1]
totalloss = 0.0

for k in range(60):
    
    for i in range(len(traindata)):
        optimizer.zero_grad()
        input = inputs[i]
        output = net(input)
        loss = criterion(output, targets[i])
        loss.backward()
        optimizer.step()
        totalloss += loss.item()
    
        if (k * len(traindata) + i) % 12000 == 0:
            print('loss: ', totalloss / 12000)
            totalloss = 0.0
```

## 线性回归

```python
point_num = 200
data = torch.Tensor([1 + 0.7 * x for x in np.linspace(0,1,point_num)]) + 0.1 * torch.randn(point_num)
print(data)
```

```python
trainPoints = pd.DataFrame({'x':np.linspace(0,1,point_num), 'y':data})
sns.relplot(x='x', y="y", data=trainPoints, aspect = 1.618)
plt.show()
plt.close()
```

```python
k = torch.Tensor([0])
b = torch.Tensor([0])

xs = torch.Tensor(np.linspace(0,1,point_num))

learnRate = 1.
change = 0.99

for i in range(2000):
    kv = torch.autograd.Variable(k, requires_grad = True)
    bv = torch.autograd.Variable(b, requires_grad = True)
    
    loss = (data - (kv * xs + bv)).norm()
    loss = loss / point_num
    loss.backward()
    
    k = k - learnRate * kv.grad
    b = b - learnRate * bv.grad
    learnRate = learnRate * change
    
print(loss / point_num)
```
